```python
miles = int(input("Please enter a value: "))
kilometers = miles * 1.60934
print("The distance in kilometers is " + str(kilometers))
print("The distance is " + str(miles) + " miles, which is the equivalent of " + str(kilometers) + " kilometers.")

```

```python

run1 = float(input("Enter the time for run 1 (in minutes): "))
run2 = float(input("Enter the time for run 2 (in minutes): "))

total_time = run1 + run2
print("Total time spent running: " + str(total_time))

improvement = run1 - run2
print("Improvement (run1 minus run2): " + str(improvement))

double_run2 = run2 * 2
print("Run2 done twice: " + str(double_run2))

average_time = (run1 + run2) / 2
print("Average time of both runs: " + str(average_time))
```

```python

quote = "  ANYTHING IS POSSIBLE! - Kevin Garnett  "
print(quote)

print("Uppercase:", quote.upper())
print("Lowercase:", quote.lower())

print("Stripped quote:", quote.strip())

print("Split into words:", quote.split())
```

```python

a = float(input("Enter the length of leg a: "))
b = float(input("Enter the length of leg b: "))

c = (a**2 + b**2) ** 0.5

c_rounded = round(c, 2)

print("The hypotenuse of your triangle is " + str(c_rounded) + " units.")
```
